4 . 
Filter the dataframe with only this columns
['continent','location','date','total_cases','total_deaths','gdp_per_ca
pita','
 human_development_index'] and update the data frame

```python
updated_covid_data = covid_data.filter(items = ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
updated_covid_data
```

5. Data Cleaning
 a. Remove all duplicates observations
 b. Find missing values in all columns
 c. Remove all observations where continent column value is
missing
 Tip : using subset parameter in dropna
 d. Fill all missing values with 0

```python
updated_covid_data.drop_duplicates()
```

```python
updated_covid_data.isnull().sum()
```

```python
updated_covid_data = updated_covid_data[updated_covid_data['continent'].notna()]
updated_covid_data
```

```python
updated_covid_data.fillna(0)
```

```python
import numpy as np

updated_covid_data['total_cases'] = updated_covid_data['total_cases'].replace(np.nan,0)
updated_covid_data['total_deaths'] = updated_covid_data['total_deaths'].replace(np.nan,0)
updated_covid_data['gdp_per_capita'] = updated_covid_data['gdp_per_capita'].replace(np.nan,0)
updated_covid_data['human_development_index'] = updated_covid_data['human_development_index'].replace(np.nan,0)
updated_covid_data
```

```python
updated_covid_data.isnull().sum()
```

6. Date time format :
 a. Convert date column in datetime format using
pandas.to_datetime
 b. Create new column month after extracting month data from
date
 column

```python
updated_covid_data['date'] = pd.to_datetime(updated_covid_data['date'])
updated_covid_data
```

```python
updated_covid_data['month'] =  pd.DatetimeIndex(updated_covid_data['date']). month
```

```python
updated_covid_data
```

7. Data Aggregation:
 a. Find max value in all columns using groupby function on
'continent'
 column
 Tip: use reset_index() after applying groupby
 b. Store the result in a new dataframe named 'df_groupby'.
 (Use df_groupby dataframe for all further analysis)

```python
df_groupby = updated_covid_data.groupby('continent').max()
```

```python
updated_covid_data.reset_index(inplace = True, drop = True)
```

```python
updated_covid_data
```

```python
df_groupby
```

8. Feature Engineering :
 a. Create a new feature 'total_deaths_to_total_cases' by ratio of
 'total_deaths' column to 'total_cases'


```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
```
